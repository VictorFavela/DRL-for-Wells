```python
import random
import time
```

```python
## Big Env
startinfo = [0,0,0]

liquidField = []

for z in range (0,100):
    plane = []
    
    for y in range (0, 100):
        line = []
        
        for x in range (0, 100):
            oil = (random.random() * .8) + .05
            water = 1 - oil
            
            oil -= .05
            water -= .05
            
            point = [oil,water]
            
            line.append(point)
        
        plane.append(line)
        
    liquidField.append(plane)            
```

```python
bigEnv = gym.make('oilfield-v1')
```

```python
bigEnv.initData(startinfo, 100, 100, 100, liquidField)
```

```python

start = time.time()

for i in range (0, 100):
    bigEnv.step(random.randrange(0,6,1))
    
end = time.time()


print("Time elapsed " + str( end - start ))
    
    
```

```python
bigEnv.render()
```
