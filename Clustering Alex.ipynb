```python
##relies on well class
#31112 is stupid name for well class
%run 31112.ipynb
import numpy as np
import matplotlib.pyplot as plt
# Though the following import is not directly being used, it is required
# for 3D projection to work
from mpl_toolkits.mplot3d import Axes3D

from sklearn.cluster import KMeans
from sklearn import datasets
```

```python
avgEasting=[]
avgNorthing=[]
avgDepth=[]
avgProppant=[]
avgPumpRate=[]

#target var
avgProduction=[]


for spot in range(29):
    avgEasting.append(np.mean(easting[spot]))
    avgNorthing.append(np.mean(northing[spot]))
    avgDepth.append(np.mean(depth[spot]))
    avgProppant.append(np.mean(proppant[spot]))
    avgPumpRate.append(np.mean(pumpRate[spot]))
for spot in range(29):
    avgProduction.append(np.mean(production[spot]))
```

```python
X=[]
y=[]
for spot in range(len(avgEasting)):
    X.append([avgEasting[spot],avgNorthing[spot],avgDepth[spot],
                  avgProppant[spot],avgPumpRate[spot]])

X=np.array(X)
y=np.array(avgProduction)
```

```python
y.shape
```

```python
estimators = [('k_Means=2', KMeans(n_clusters=2)),
              ('k_Means=3', KMeans(n_clusters=3)),
              ('k_Means=5', KMeans(n_clusters=5))]
titles = ['2 clusters', '3 clusters', '5 clusters']

fignum = 1

for name, est in estimators:
    fig = plt.figure(fignum, figsize=(4, 3))
    ax = Axes3D(fig, rect=[0, 0, .95, 1], elev=48, azim=134)
    est.fit(X)
    labels = est.labels_

    ax.scatter(X[:, 0], X[:, 1], X[:, 2],
               c=labels.astype(np.float), edgecolor='k')

    ax.w_xaxis.set_ticklabels([])
    ax.w_yaxis.set_ticklabels([])
    ax.w_zaxis.set_ticklabels([])
    ax.set_xlabel('avgEasting')
    ax.set_ylabel('avgNorthing')
    ax.set_zlabel('avgDepth')
    ax.set_title(titles[fignum - 1])
    ax.dist = 12
    fignum = fignum + 1
```
